```python
def find_daterange(quarter):
    print('Dates between '+ df_garment_prod[df_garment_prod['quarter'] == quarter]['date'].min().strftime('%m/%d/%Y') + 
          ' and ' + df_garment_prod[df_garment_prod['quarter'] == quarter]['date'].max().strftime('%m/%d/%Y') +
          ' in ' + quarter)
    
find_daterange('Quarter1')
find_daterange('Quarter2')
find_daterange('Quarter3')
find_daterange('Quarter4')
find_daterange('Quarter5') 
```

## 8.4 How does productivity of workers vary through the month of Jan 2015 to Mar 2015?

```python
fig, ax = plt.subplots(figsize=(10, 6))
df_garment_actual_prod = df_garment_prod.groupby(['date'])['actual_productivity'].mean().sort_index()
df_garment_actual_prod = df_garment_actual_prod.reset_index()
df_garment_target_prod = df_garment_prod.groupby(['date'])['targeted_productivity'].mean().sort_index()
df_garment_target_prod = df_garment_target_prod.reset_index()
ax.plot( 'date', 'targeted_productivity', data=df_garment_target_prod, marker='o', markerfacecolor='blue', color='skyblue', linewidth=3, label="Target")
ax.plot( 'date', 'actual_productivity', data=df_garment_actual_prod, marker='o', markerfacecolor='red', color='red', linewidth=3, label="Actual" )
    
    # rotate_labels...
for label in ax.get_xticklabels():
    label.set_rotation(40)
    label.set_horizontalalignment('right')
plt.legend()
plt.xlabel("Date")
plt.ylabel("Productivity")
plt.title("Actual and Target Productivity Vs Date ")
```

**Observations:**


- **From the beginning of Jan 2015 until 15-Jan-2015**, the actual productivity is pretty consistent when compared to targeted productivity. 
- Then then is a sharp increase in productivity from **mid of Jan 2015 to mid of Feb 2015**. 
- The trend decreases in the **later half of Feb 2016** when compared to targeted productivity.

__Linear regression__ is a _basic_ and _commonly_ used type of __predictive analysis__.  The overall idea of regression is to examine two things: 
- Does a set of __predictor variables__ do a good job in predicting an __outcome__ (dependent) variable?  
- Which variables in particular are __significant predictors__ of the outcome variable, and in what way they do __impact__ the outcome variable?  

These regression estimates are used to explain the __relationship between one dependent variable and one or more independent variables__.  The simplest form of the regression equation with one dependent and one independent variable is defined by the formula :<br/>
$y = \beta_0 + \beta_1x$

### 10.2.7  Model Evaluation

__Error__ is the _deviation_ of the values _predicted_ by the model with the _true_ values.<br/>

Below are the types of error we will be calculating for our _linear regression model_:
- Mean Absolute Error
- Mean Squared Error
- Root Mean Squared Error

```python
MAE_train = mean_absolute_error(y_train, y_pred_train)
MAE_test = mean_absolute_error(y_test, y_pred_test)
RMSE_train = np.sqrt( mean_squared_error(y_train, y_pred_train))
RMSE_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print('MAE for training set is {}'.format(MAE_train))
print('MAE for test set is {}'.format(MAE_test))
print('RMSE for training set is {}'.format(RMSE_train))
print('RMSE for test set is {}'.format(RMSE_test))
```

There is one more method to evaluate linear regression model and that is by using the Rsquared value.

R-squared is the proportion of variance explained, meaning the proportion of variance in the observed data that is explained by the model, or the reduction in error over the null model. (The null model just predicts the mean of the observed response, and thus it has an intercept and no slope.)

R-squared is between 0 and 1, and higher is better because it means that more variance is explained by the model. But there is one shortcoming of Rsquare method and that is R-squared will always increase as you add more features to the model, even if they are unrelated to the response. Thus, selecting the model with the highest R-squared is not a reliable approach for choosing the best linear model.

There is alternative to R-squared called adjusted R-squared that penalizes model complexity (to control for overfitting).

```python
r2_score_train = r2_score(y_train, y_pred_train)
r2_score_test = r2_score(y_test, y_pred_test)
print('R squared score for training set is {}'.format(r2_score_train))
print('R squared score for test set is {}'.format(r2_score_test))
```

```python
yhat = regressor.predict(X_train)
SS_Residual = sum((y_train-yhat)**2)
SS_Total = sum((y_train-np.mean(y_train))**2)
r_squared = 1 - (float(SS_Residual))/SS_Total
adj_r_sq_train = 1 - (1-r_squared)*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1)
print('Adjusted R squared score for training set is {}'.format(adj_r_sq_train))

yhat = regressor.predict(X_test)
SS_Residual = sum((y_test-yhat)**2)
SS_Total = sum((y_test-np.mean(y_test))**2)
r_squared = 1 - (float(SS_Residual))/SS_Total
adj_r_sq_test = 1 - (1-r_squared)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print('Adjusted R squared score for testing set is {}'.format(adj_r_sq_test))
```

```python
#loop the value of alpha for different tree
clf_dts = []
for ccp_alpha  in ccp_alphas:
    dt_garment_prod_ccp = DecisionTreeClassifier(random_state=42 , ccp_alpha=ccp_alpha )
    dt_garment_prod_ccp.fit(X_train , y_train)
    clf_dts.append(dt_garment_prod_ccp)
```

Now we will plot the graph between **accuracy** and  **different data set alpha values** calculated above

```python
train_scores = [dt_garment_prod_ccp.score(X_train , y_train) for dt_garment_prod_ccp in clf_dts ]
test_scores = [dt_garment_prod_ccp.score(X_test , y_test) for dt_garment_prod_ccp in clf_dts ]


fig , ax = plt.subplots()
ax.set_xlabel = "Accuracy"
ax.set_ylabel = "Alphas"
ax.set_title = "Accuracy vs Alphas for testing and training datasets"
ax.plot(ccp_alphas , train_scores , marker = 'o' , label = "Train" ,  drawstyle = "steps-post")
ax.plot(ccp_alphas , test_scores , marker = 'o' , label = "Test" ,  drawstyle = "steps-post" )
ax.legend()
plt.show()
```

**Accuracy** of test dataset is maximum when the **alpha** value is aroung **0.004**

In **cost complexity pruning** the value of **alpha** ranges from **0 to 1** as **Gini** varies from **0 to 1** .

In theory , **alpha** can go much larger as **sum of squared residuals** range from **0 to positive infinity**.

How do we know that we have taken a best  **training and testing** dataset?? . This can be achieved by **k-fold cross validation**.


Now we do check the accuracy by using the value of **alpha = 0.004** for the training and testing dataset for which we claculated alpha


```python
dt_garment_prod_ccp = DecisionTreeClassifier(random_state=42 , ccp_alpha=0.004 )

# 5 fold cross validation with aplha = 0.016
cv_score= cross_val_score(dt_garment_prod_ccp ,  X_train , y_train , cv=10)

# build a dataframe of 5 trees and scores
df = pd.DataFrame( {"tree": range(10), "accuracy" :cv_score })
df.plot(x= "tree" , y="accuracy" , marker='o' , linestyle = '--' )
```

It gives the accuracy at one point but later it dips. It gives very bad accuracy at a point. It suggests that alpha is sensitive to dataset . We have to do cross validation of the dataset and then calculate different alpha values

```python
#loop the value of alpha for different tree
alpha_value_loops = []

# for each alpha value we will do 5 -fold cross validation 
for ccp_alpha  in ccp_alphas:
    dt_garment_prod_ccp = DecisionTreeClassifier(random_state=42 , ccp_alpha=ccp_alpha )
    cv_score= cross_val_score(dt_garment_prod_ccp ,  X_train , y_train , cv=10)
    alpha_value_loops.append([ccp_alpha , np.mean(cv_score) , np.std(cv_score)])
    
alpha_value_results = pd.DataFrame(alpha_value_loops , columns =["alpha","mean_accuracy" , "std_deviation"] )

alpha_value_results.plot(x="alpha" , y="mean_accuracy" , yerr="std_deviation" , marker="o" , linestyle="--" , )
```
